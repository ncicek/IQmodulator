```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
wave = pd.read_csv("../tb/cocotb/dds_sim/wave.csv").drop(['Unnamed: 0'], axis=1)
wave
```

```python
wave[0:10000].plot(y=['phase_i', 'phase_q'])
```

```python
plt.plot(signal.signaltools.correlate(wave['wave_i'], wave['wave_i']))
plt.xlim(1e6-5,1e6+5)
plt.ylim(2.1486e15,2.149e15)
```

```python
wave['phase_diff'] = (wave['phase_i'] - wave['phase_q'])
```

```python
wave['phase_diff'][0:1000].plot()
```

```python
plt.magnitude_spectrum(wave['wave_q'], color ='green', Fs=100e6, scale='dB') 
plt.ylim(-70,100)
plt.xlim(0,1e6)
```

```python
plt.magnitude_spectrum(wave['wave_i'], color ='green', Fs=100e6, scale='dB') 
plt.ylim(-70,100)
plt.xlim(0,1e6)
```

```python
plt.phase_spectrum(wave['wave_q'], color ='green', Fs=100e6) 
```
